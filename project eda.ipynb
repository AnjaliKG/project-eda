# State

```python
d.groupby('State')['Customer Lifetime Value'].mean()
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,5))
d.State.value_counts().plot.pie(shadow=True,autopct='%1.1f%%',ax=ax[0],colors =['teal','lightseagreen','mediumaquamarine','palegreen','cyan'])

sns.barplot(x='State',y="Customer Lifetime Value",palette="Paired", edgecolor=".6",data=d,ax=ax[1])
plt.title('States vs Customer Lifetime Value ')
plt.show()
```

**1. 34.5% of customers are from California.**

**2. Customers from Washington are comparitively less in number.**

**3. Average customer lifetime value is low for the customers from Arizona.**

**4. Customers from all other states have almost equal average customer lifetime value.**

#  Coverage

```python
d.groupby('Coverage')['Customer Lifetime Value'].mean()
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,5))
d.Coverage.value_counts(normalize=True).plot.pie(shadow=True,autopct='%1.1f%%',ax=ax[0],colors =['teal','mediumaquamarine','palegreen'])
plt.title('Coverage vs count')


sns.barplot(x='Coverage',y="Customer Lifetime Value",palette="viridis", edgecolor=".6",data=d,ax=ax[1])
plt.title('Coverage vs Customer Lifetime Value ')
plt.show()
```

**1. 61% of customers hold  basic coverage.But average customer lifetime value is low for customers with basic coverage.**

**2. Customers with premium coverage higher Customer lifetime value eventhough customers holding premium coverage are less in count.**

# Response

```python
d.groupby('Response')['Customer Lifetime Value'].mean()
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,5))
axi=sns.countplot(x='Response',palette="viridis",data=d,ax=ax[0])
plt.title('Response')
for p in axi.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/float(len(d)))
    x = (p.get_x() + p.get_width())/1.23
    y = p.get_height()
    axi.annotate(percentage, (x, y),ha='center')

   

sns.barplot(x='Response',y="Customer Lifetime Value",palette="viridis", edgecolor=".6",data=d,ax=ax[1])
plt.title('Response vs Customer Lifetime Value')
```

**1. 85.7% of customers's don't like to renew their insurance .**

**2. Average customer lifetime value is slightly higher for customers who don't like to renew their insurance. So it is important to retain those customers .**

# Policy Type And Number Of Policies

```python
d.groupby(d['Policy Type'])['Customer Lifetime Value'].mean()
```

```python
fig,ax=plt.subplots(1,2,figsize=(15,8))

sns.countplot(x='Number of Policies',data=d,ax=ax[0],palette='Paired')
plt.title('Number of Policies')

sns.countplot(x='Policy Type',data=d,ax=ax[1],palette='viridis')
plt.title('Policy Type')

axi=sns.catplot(x='Number of Policies',y="Customer Lifetime Value", hue="Policy Type", kind="bar",
            palette="viridis",height=8, aspect=15/8,
            data=d)
plt.title('Number Of Policies vs Customer Lifetime Value')
plt.show()
```

**3. 74.3% .customers hold personal auto policy while only 4% hold special auto policy.**

**4. Customer lifetime value is high for the customers with 2 policies .**

**5. Customers with Special auto policy has the highest customer lifetime value while the least is with corporate auto policy.**

**6. Single policy holders have comparatively less customer lifetime value.**
